```py
    class Animal:
        def __init__(self, name):
            self.name = name
        
        def speak(self):
            return "Some sound"

    class Dog(Animal):
        def speak(self):
            return "Woof"

    class Cat(Animal):
        def speak(self):
            return super().speak() + " meow"

    animals = [Animal("Generic"), Dog("Buddy"), Cat("Whiskers")]
    for animal in animals:
        print(animal.speak())
```
